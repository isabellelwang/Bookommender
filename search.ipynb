```python
def parse_line(line): 
    '''Parse each line to extract important information about a book'''

    df = json.loads(line)

    return {
        'book_id' : df['book_id'], 
        'tile' : df['title_without_series'], 
        'ratings' : df['ratings_count'],
        'url' : df['url'], 
        'cover_image' : df['image_url'] 
    }

    
```

```python
parse_line(line)
```

```python
book_titles = []
with gzip.open("goodreads_books.json.gz", 'r') as f:
    while True:  
        line = f.readline() # read each line 

        # if the line does not contain anything, break 
        if not line: 
            break 
        
        data = parse_line(line)

        try: 
            ratings = int(data['ratings'])
        except ValueError: 
            continue

        if ratings > 15: 
            book_titles.append(data)

    
```

```python
book_titles
```

```python
import pandas as pd 

# create a dataframe for dictionary of books 

titles = pd.DataFrame.from_dict(data = book_titles)
```

```python
# change ratings into numeric 
titles['ratings'] = pd.to_numeric(titles['ratings'])
```
